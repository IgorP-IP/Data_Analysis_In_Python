```python
#  :    ()      ?
#   : case, site, Pop, sex
df = df.drop(columns=['case', 'site', 'Pop', 'sex'])
df.head(3)
```

```python
# ,     dataset 
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

  dataset   -  X -   (target)  - y

```python
X = df.drop(columns=['age']).values
y = df['age'].values
```

  dataset  trainset  test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle = True, test_size = 0.2, random_state=42)
```

```python
#  1.  
model = KNeighborsRegressor(n_neighbors = 5)

#  2.  
model.fit(X_train, y_train)

#  3.    
y_pred = model.predict(X_test)
```

### 3.

   .          .

 -  ,            , 

   -  ""  .

   : MSE, RMSE, R^2, MAE.

   : Accuracy, Precision, Recall, F1-measure, ROC AUC,  

MSE, RMSE, R^2, MAE

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score, mean_absolute_error

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

MSE_train = mean_squared_error(y_train, pred_train)
RMSE_train = np.sqrt(MSE_train)
R2_train = r2_score(y_train, pred_train)
MAE_train = mean_absolute_error(y_train, pred_train)

MSE_test = mean_squared_error(y_test, pred_test)
RMSE_test = np.sqrt(MSE_test)
R2_test = r2_score(y_test, pred_test)
MAE_test = mean_absolute_error(y_test, pred_test)

print(f'MSE   {MSE_train:.2f}')
print(f'MSE   {MSE_test:.2f}', end = '\n\n')

print(f'RMSE   {RMSE_train:.2f}')
print(f'RMSE   {RMSE_test:.2f}', end = '\n\n')

print(f'R^2   {R2_train:.2f}')
print(f'R^2   {R2_test:.2f}', end = '\n\n')

print(f'MAE   {MAE_train:.2f}')
print(f'MAE   {MAE_test:.2f}', end = '\n\n')
```

```python
#  
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state = 44)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score, mean_absolute_error

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

MSE_train = mean_squared_error(y_train, pred_train)
RMSE_train = np.sqrt(MSE_train)
R2_train = r2_score(y_train, pred_train)
MAE_train = mean_absolute_error(y_train, pred_train)

MSE_test = mean_squared_error(y_test, pred_test)
RMSE_test = np.sqrt(MSE_test)
R2_test = r2_score(y_test, pred_test)
MAE_test = mean_absolute_error(y_test, pred_test)

print(f'MSE   {MSE_train:.2f}')
print(f'MSE   {MSE_test:.2f}', end = '\n\n')

print(f'RMSE   {RMSE_train:.2f}')
print(f'RMSE   {RMSE_test:.2f}', end = '\n\n')

print(f'R^2   {R2_train:.2f}')
print(f'R^2   {R2_test:.2f}', end = '\n\n')
```
