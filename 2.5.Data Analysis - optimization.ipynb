```python
import numpy as np

#   (y)    (p)
y_true = np.array([1, 0, 1, 1, 0])  #   (0  1)
y_pred = np.array([0.9, 0.1, 0.8, 0.7, 0.2])  #     1

#    - 
def binary_crossentropy(y_true, y_pred):
    #  ,      (0, 1) 
    # (   0  1)
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    
    #  - 
    loss = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return loss

#  -    
loss = binary_crossentropy(y_true, y_pred)
print(f"- : {loss}")
```

```python
import numpy as np

#     
y_true = np.array([3.0, 2.5, 4.1, 5.0, 6.2])  #  
y_pred = np.array([2.8, 2.7, 4.0, 5.1, 6.0])  #  

#      (MSE)
def mean_squared_error(y_true, y_pred):
    #   
    mse = np.mean((y_true - y_pred) ** 2)
    return mse

#      
mse_value = mean_squared_error(y_true, y_pred)
print(f"  (MSE): {mse_value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  f(x) = x^2 + 6x + 9
def f(x):
    return x**2 + 6*x + 9

#   f(x),   
def df(x):
    return 2*x + 6

#  
def gradient_descent(learning_rate, epochs, initial_x):
    x = initial_x
    x_values = [x]  #    x   
    for epoch in range(epochs):
        grad = df(x)  #  
        x = x - learning_rate * grad  #  x
        x_values.append(x)  #    x
    return x, x_values

# 
learning_rate = 0.1  #  
epochs = 50  #  
initial_x = 5  #  

#   
min_x, x_values = gradient_descent(learning_rate, epochs, initial_x)

#  
print(f"     x = {min_x}")

#  
x_vals = np.linspace(-10, 2, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x^2 + 6x + 9", color='b')  #  
plt.scatter(x_values, f(np.array(x_values)), color='r', label="  ", zorder=5)  #  
plt.plot(x_values, f(np.array(x_values)), color='r', linestyle='--', alpha=0.5)  # 
plt.title(" ")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  f(x) = x^2 + 6x + 9
def f(x):
    return x**2 + 6*x + 9

#   f(x),   
def df(x):
    return 2*x + 6

#   
def stochastic_gradient_descent(learning_rate, epochs, initial_x):
    x = initial_x
    x_values = [x]  #    x   
    for epoch in range(epochs):
        #    (       )
        grad = df(x)  #  
        x = x - learning_rate * grad  #  x
        x_values.append(x)  #    x
    return x, x_values

# 
learning_rate = 0.1  #  
epochs = 50  #  
initial_x = 5  #  

#    
min_x, x_values = stochastic_gradient_descent(learning_rate, epochs, initial_x)

#  
print(f"     x = {min_x}")

#  
x_vals = np.linspace(-10, 2, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x^2 + 6x + 9", color='b')  #  
plt.scatter(x_values, f(np.array(x_values)), color='r', label="   ", zorder=5)  #  
plt.plot(x_values, f(np.array(x_values)), color='r', linestyle='--', alpha=0.5)  # 
plt.title("  ")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

###  5: L1-




```python
import numpy as np
import matplotlib.pyplot as plt

#      
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100 , 1 
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3X + 

#    L1-
def loss_with_l1(X, y, theta, lambda_reg):
    m = len(y)
    predictions = X.dot(theta)  # 
    loss = (1 / (2 * m)) * np.sum((predictions - y) ** 2)  #  
    regularization = lambda_reg * np.sum(np.abs(theta))  # L1-
    total_loss = loss + regularization  #   
    return total_loss

#     L1-
def gradient_with_l1(X, y, theta, lambda_reg):
    m = len(y)
    predictions = X.dot(theta)  # 
    gradient = (1 / m) * X.T.dot(predictions - y)  #  
    regularization_grad = lambda_reg * np.sign(theta)  #  L1-
    total_gradient = gradient + regularization_grad  #  
    return total_gradient

#    L1-
def gradient_descent_l1(X, y, theta_initial, lambda_reg, learning_rate, iterations):
    theta = theta_initial
    loss_history = []
    
    for i in range(iterations):
        gradient = gradient_with_l1(X, y, theta, lambda_reg)
        theta = theta - learning_rate * gradient  #  
        loss = loss_with_l1(X, y, theta, lambda_reg)
        loss_history.append(loss)  #     
    
    return theta, loss_history

# 
lambda_reg = 0.1  #  
learning_rate = 0.1  #  
iterations = 1000  #  
theta_initial = np.random.randn(2, 1)  #     

#     X    ( )
X_b = np.c_[np.ones((X.shape[0], 1)), X]

#     L1-
theta_final, loss_history = gradient_descent_l1(X_b, y, theta_initial, lambda_reg, learning_rate, iterations)

#  
print(f" : {theta_final}")

#     
plt.plot(loss_history)
plt.title('    L1-')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid(True)
plt.show()

#     
plt.scatter(X, y, color='blue', label='')
plt.plot(X, X_b.dot(theta_final), color='red', label='   L1-')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

###  6: L2-



```python
import numpy as np
import matplotlib.pyplot as plt

#      
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100 , 1 
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3X + 

#    L2-
def loss_with_l2(X, y, theta, lambda_reg):
    m = len(y)
    predictions = X.dot(theta)  # 
    loss = (1 / (2 * m)) * np.sum((predictions - y) ** 2)  #  
    regularization = (lambda_reg / 2) * np.sum(theta[1:] ** 2)  # L2- (  )
    total_loss = loss + regularization  #   
    return total_loss

#     L2-
def gradient_with_l2(X, y, theta, lambda_reg):
    m = len(y)
    predictions = X.dot(theta)  # 
    gradient = (1 / m) * X.T.dot(predictions - y)  #  
    regularization_grad = lambda_reg * theta  #  L2-
    regularization_grad[0] = 0  #    ( )
    total_gradient = gradient + regularization_grad  #  
    return total_gradient

#    L2-
def gradient_descent_l2(X, y, theta_initial, lambda_reg, learning_rate, iterations):
    theta = theta_initial
    loss_history = []
    
    for i in range(iterations):
        gradient = gradient_with_l2(X, y, theta, lambda_reg)
        theta = theta - learning_rate * gradient  #  
        loss = loss_with_l2(X, y, theta, lambda_reg)
        loss_history.append(loss)  #     
    
    return theta, loss_history

# 
lambda_reg = 1.0  #  
learning_rate = 0.1  #  
iterations = 1000  #  
theta_initial = np.random.randn(2, 1)  #     

#     X    ( )
X_b = np.c_[np.ones((X.shape[0], 1)), X]

#     L2-
theta_final, loss_history = gradient_descent_l2(X_b, y, theta_initial, lambda_reg, learning_rate, iterations)

#  
print(f" : {theta_final}")

#     
plt.plot(loss_history)
plt.title('    L2-')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid(True)
plt.show()

#     
plt.scatter(X, y, color='blue', label='')
plt.plot(X, X_b.dot(theta_final), color='red', label='   L2-')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

#    MAE
def mean_absolute_error(y_true, y_pred):
    # y_true -  , y_pred -  
    return np.mean(np.abs(y_true - y_pred))

#  
y_true = np.array([3.0, -0.5, 2.0, 7.0])  #  
y_pred = np.array([2.5, 0.0, 2.1, 7.8])  #  

#  MAE
mae = mean_absolute_error(y_true, y_pred)

print(f"Mean Absolute Error: {mae}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  (x^2)
def function(x):
    return x ** 2

#   (2x)
def gradient(x):
    return 2 * x

#  
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    history = [x]  #    

    for _ in range(iterations):
        grad = gradient(x)  #   
        x = x - learning_rate * grad  #   x
        history.append(x)  #    x  

    return x, history

#   
starting_point = 5  #  
learning_rate = 0.1  #  
iterations = 50  #  

#   
minimized_value, history = gradient_descent(starting_point, learning_rate, iterations)

#    
print(f"  : {minimized_value}")
print(f" : {iterations}")

#  
x_vals = np.linspace(-6, 6, 100)  #     x
y_vals = function(x_vals)  #    

plt.plot(x_vals, y_vals, label="f(x) = x^2", color='blue')
plt.scatter(history, function(np.array(history)), color='red', label=' ')
plt.plot(history, function(np.array(history)), linestyle="--", color='red')
plt.title("    x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
