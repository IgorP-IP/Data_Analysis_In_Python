```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 400)
y = x**2
dy = 2*x

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x)=x', color='blue')
plt.plot(x, dy, label="f'(x)=2x", color='red')
plt.title('    ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

dz_dx = 2*x
dz_dy = 2*y

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.7)
ax.quiver(x, y, z, dz_dx, dz_dy, 0, color='red')
ax.set_title('    ')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, solve

# 1.  
x = symbols('x')
f = x**3 - 6*x**2 + 9*x

# 2.   
f_prime = diff(f, x)
print(f"  f'(x) = {f_prime}")

# 3.    (   )
critical_points = solve(f_prime, x)
print(f"  ( f'(x) = 0): {critical_points}")

# 4.      

#         
f_num = lambda x: x**3 - 6*x**2 + 9*x
f_prime_num = lambda x: 3*x**2 - 12*x + 9

#    x
x_vals = np.linspace(-2, 6, 400)

#      
y_vals = f_num(x_vals)
y_prime_vals = f_prime_num(x_vals)

#  
plt.figure(figsize=(10, 6))

#   f(x)
plt.plot(x_vals, y_vals, label=r'$f(x) = x^3 - 6x^2 + 9x$', color='b')

#   f'(x)
plt.plot(x_vals, y_prime_vals, label=r"$f'(x) = 3x^2 - 12x + 9$", color='r', linestyle='--')

#   
for point in critical_points:
    if point.is_real:  #   
        plt.scatter(float(point), f_num(float(point)), color='g', zorder=5)
        plt.text(float(point), f_num(float(point)), f'({float(point):.2f}, {f_num(float(point)):.2f})', fontsize=12, color='g')

#  
plt.title('    ')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff

#  
x, y = symbols('x y')

#   f(x, y)
f = x**2 + 2*x*y + y**2

#     x  y
df_dx = diff(f, x)
df_dy = diff(f, y)

print(f"   x: {df_dx}")
print(f"   y: {df_dy}")

#         
f_num = lambda x, y: x**2 + 2*x*y + y**2
df_dx_num = lambda x, y: 2*x + 2*y
df_dy_num = lambda x, y: 2*x + 2*y

#     x  y
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)

#         
X, Y = np.meshgrid(x_vals, y_vals)
Z = f_num(X, Y)
Z_dx = df_dx_num(X, Y)
Z_dy = df_dy_num(X, Y)

#    f(x, y)
plt.figure(figsize=(14, 5))

#   f(x, y)
plt.subplot(1, 3, 1)
cp = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(cp)
plt.title(r"$f(x, y) = x^2 + 2xy + y^2$")
plt.xlabel('x')
plt.ylabel('y')

#     x
plt.subplot(1, 3, 2)
cp = plt.contourf(X, Y, Z_dx, levels=50, cmap='coolwarm')
plt.colorbar(cp)
plt.title(r"$\frac{\partial f}{\partial x}$")
plt.xlabel('x')
plt.ylabel('y')

#     y
plt.subplot(1, 3, 3)
cp = plt.contourf(X, Y, Z_dy, levels=50, cmap='coolwarm')
plt.colorbar(cp)
plt.title(r"$\frac{\partial f}{\partial y}$")
plt.xlabel('x')
plt.ylabel('y')

#   
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff

#  
x, y = symbols('x y')

#   f(x, y)
f = x**2 + 2*x*y + y**2

#     x  y ()
df_dx = diff(f, x)
df_dy = diff(f, y)

#  
gradient = np.array([df_dx, df_dy])
print(f"  f(x, y): {gradient}")

#        
f_num = lambda x, y: x**2 + 2*x*y + y**2
df_dx_num = lambda x, y: 2*x + 2*y
df_dy_num = lambda x, y: 2*x + 2*y

#     x  y
x_vals = np.linspace(-5, 5, 20)
y_vals = np.linspace(-5, 5, 20)

#          
X, Y = np.meshgrid(x_vals, y_vals)
Z = f_num(X, Y)
Z_dx = df_dx_num(X, Y)
Z_dy = df_dy_num(X, Y)

#    f(x, y)   

plt.figure(figsize=(12, 6))

#   f(x, y) ()
plt.subplot(1, 2, 1)
cp = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(cp)
plt.title(r"$f(x, y) = x^2 + 2xy + y^2$")
plt.xlabel('x')
plt.ylabel('y')

#  
plt.subplot(1, 2, 2)
plt.quiver(X, Y, Z_dx, Z_dy, color='r', scale=50)
plt.title(" ")
plt.xlabel('x')
plt.ylabel('y')

#   
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, lambdify

#   x
x = symbols('x')

#  
f = x**3 - 4*x + 2

#  
f_prime = diff(f, x)
print(f" : f'(x) = {f_prime}")

#      
f_func = lambdify(x, f, 'numpy')
f_prime_func = lambdify(x, f_prime, 'numpy')

#      
x_vals = np.linspace(-3, 3, 400)
y_vals = f_func(x_vals)
y_prime_vals = f_prime_func(x_vals)

#      
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='f(x)=x-4x+2', color='blue')
plt.plot(x_vals, y_prime_vals, label="f'(x)=3x-4", color='red', linestyle='--')
plt.title('    ')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.legend()
plt.grid(True)
plt.show()
```

### :       scipy



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

#  
def f(x):
    return x**2

#     0  3
integral, error = quad(f, 0, 3)

print(f"  f(x) = x^2  0  3: {integral:.4f}")

#   x   
x_vals = np.linspace(0, 3, 100)
y_vals = f(x_vals)

#     x ( )
integral_vals = np.array([quad(f, 0, x)[0] for x in x_vals])

#  
plt.figure(figsize=(10, 6))

#   f(x)
plt.subplot(2, 1, 1)
plt.plot(x_vals, y_vals, label=r'$f(x) = x^2$', color='blue')
plt.title('  f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

#  
plt.subplot(2, 1, 2)
plt.plot(x_vals, integral_vals, label=' ', color='green')
plt.title('   f(x) = x^2')
plt.xlabel('x')
plt.ylabel('')
plt.legend()

#  
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

#   f(x) = x^3 - 2x + 1
def f(x):
    return x**3 - 2*x + 1

#     0  2
integral, error = quad(f, 0, 2)

print(f"  f(x) = x^3 - 2x + 1  0  2: {integral:.4f}")

#   x   
x_vals = np.linspace(0, 2, 100)
y_vals = f(x_vals)

#     x ( )
integral_vals = np.array([quad(f, 0, x)[0] for x in x_vals])

#  
plt.figure(figsize=(10, 6))

#   f(x)
plt.subplot(2, 1, 1)
plt.plot(x_vals, y_vals, label=r'$f(x) = x^3 - 2x + 1$', color='blue')
plt.title('  f(x) = x^3 - 2x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

#  
plt.subplot(2, 1, 2)
plt.plot(x_vals, integral_vals, label=' ', color='green')
plt.title('   f(x) = x^3 - 2x + 1')
plt.xlabel('x')
plt.ylabel('')
plt.legend()

#  
plt.tight_layout()
plt.show()
```
